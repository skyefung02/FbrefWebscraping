```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 

html_salah_summary_2324 = requests.get('https://fbref.com/en/players/e342ad68/matchlogs/2023-2024/c9/Mohamed-Salah-Match-Logs').text
soup_summary = BeautifulSoup(html_salah_summary_2324, 'lxml')
salah_data_summary = soup_summary.find('tbody')
```

```python
#getting data function 
def get_dat(dat, col):
    coldat = []
    for element in dat.find_all('td', attrs ={'data-stat': col}):
        coldat.append(element.text)
    coldat = [item for item in coldat if item != '']
    return (coldat)    
```

```python
#getting the data for the necessary columns

#misc data 
date = get_dat(salah_data_summary,'dayofweek')
venue = get_dat(salah_data_summary, 'venue')
result = get_dat(salah_data_summary, 'result')
team = get_dat(salah_data_summary, 'team')
opponent = get_dat(salah_data_summary, 'opponent')
start = get_dat(salah_data_summary, 'game_started')
position = get_dat(salah_data_summary, 'position')
mins = get_dat(salah_data_summary, 'minutes')

#performance data
goals = get_dat(salah_data_summary, 'goals')
assist = get_dat(salah_data_summary, 'assists')
pen_goals = get_dat(salah_data_summary, 'pens_made')
pen_attempted = get_dat(salah_data_summary, 'pens_att')
shots = get_dat(salah_data_summary, 'shots')
sot = get_dat(salah_data_summary, 'shots_on_target')
yellow = get_dat(salah_data_summary, 'cards_yellow')
red = get_dat(salah_data_summary, 'cards_red')
touches = get_dat(salah_data_summary, 'touches')
tackles = get_dat(salah_data_summary, 'tackles')
interceptions = get_dat(salah_data_summary, 'interceptions')
blocks = get_dat(salah_data_summary, 'blocks')

#expected performance metrics
xg = get_dat(salah_data_summary, 'xg')
npxg = get_dat(salah_data_summary, 'npxg')
xag = get_dat(salah_data_summary, 'xg_assist')

#shot creating actions
sca = get_dat(salah_data_summary, 'sca')
gca = get_dat(salah_data_summary, 'gca')

#passes
passes_completed = get_dat(salah_data_summary, 'passes_completed')
passes_attempted = get_dat(salah_data_summary, 'passes')
pass_percentage = get_dat(salah_data_summary, 'passes_pct')
prg_passes = get_dat(salah_data_summary, 'progressive_passes')

#carries 
carries = get_dat(salah_data_summary, 'carries')
prg_carries = get_dat(salah_data_summary, 'progressive_carries')

#take-ons
takeon_att = get_dat(salah_data_summary, 'take_ons')
takeon_suc = get_dat(salah_data_summary, 'take_ons_won')
```

```python
#putting all of the arrays into a pd dataframe
salah2324_df = df = pd.DataFrame({
    'Date': date,
    'Venue': venue,
    'Result': result,
    'Team': team,
    'Opponent': opponent, 
    'Start': start, 
    'Position': position,
    'Minutes Played': mins, 
    'Goals': goals, 
    'Assists': assist, 
    'Penalties Scored': pen_goals, 
    'Penalties Attempted': pen_attempted, 
    'Shots': shots, 
    'Shots on Target': sot, 
    'Yellow Cards': yellow, 
    'Red Cards': red, 
    'Touches': touches, 
    'Tackles': tackles,
    'Interceptions': interceptions, 
    'Blocks': blocks, 
    'xG': xg,
    'npxG': npxg,
    'xAG': xag, 
    'Shot Creating Actions': sca, 
    'Goal Creating Actions': gca, 
    'Passes Completed': passes_completed, 
    'Passes Attempted': passes_attempted, 
    'Pass Completion Percentage': pass_percentage,
    'Progressive Passes': prg_passes,
    'Carries': carries, 
    'Progressive Carries': prg_carries, 
    'Take-ons Attempted': takeon_att, 
    'Successful Take-ons': takeon_suc
})
```
